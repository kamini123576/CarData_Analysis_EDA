**Step 3: Ensure Numerical Columns are Correctly Formatted and Handle Outliers.**

**1. Check for outliers using the Interquartile Range (IQR) method.**

```python
# Defining a Function to check for outliers using IQR
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Cap outliers to the lower and upper bounds
    df[column] = np.where(df[column] < lower_bound, lower_bound, df[column])
    df[column] = np.where(df[column] > upper_bound, upper_bound, df[column])
```

```python
# Apply outlier removal to all numerical columns
for col in numerical_columns:
    remove_outliers(data, col)

# Display summary statistics after cleaning
print("\nSummary Statistics After Outlier Handling:")
print(data[numerical_columns].describe())

```

**4.** **Generating tags/features from free text available :**

**a. Import Necessary Libraries**

```python
import pandas as pd
import nltk
from collections import Counter
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Combine free-text columns into a single column
data['COMBINED_TEXT'] = data[['CUSTOMER_VERBATIM', 'CORRECTION_VERBATIM']].fillna('').agg(' '.join, axis=1)
```

**b. Define a function to clean and tokenize text.**

```python
# Remove non-alphabetic characters and convert to lowercase
import re
def clean_and_tokenize(text):
    tokens = re.findall(r'\b[a-z]+\b', text.lower())
    return tokens

data['TOKENS'] = data['COMBINED_TEXT'].apply(clean_and_tokenize)
```

**c. Generate Tags Using Word Frequency**

```python
all_tokens = [token for tokens in data['TOKENS'] for token in tokens]

# Count word frequencies
word_freq = Counter(all_tokens).most_common(10)

# Extract the top tags
tags = [word for word, freq in word_freq]
print("Top Tags:", tags)

# Add tags
data['TAGS'] = data['TOKENS'].apply(lambda tokens: ', '.join([word for word in tags if word in tokens]))
```
